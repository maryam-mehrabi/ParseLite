# **LSH Sampling**

**MinHash LSH**

```python
lsh = MinHashLSH(threshold=0.6, num_perm=64) ##### (3) Definition of hash table. Hashing and puting signatures to buckets. It will be done internally. The length of the signatures generated by MinHashing are 128 integers

minhashes = []
for i, log in enumerate(logs):
    m = MinHash(num_perm=64) ##### MinHash object will create the signatures for each set of shingles
    for token in log.split():
        m.update(token.encode('utf8')) ##### (1) Shingling step: turn the log message to tokens. Here the shingles are tokens separated by space
    lsh.insert(f"log_{i}", m) ##### (2) Signature generation for each shingle set  and it will be inserted to hash table buckets
    minhashes.append(m)

print("LSH index created.")
```

**Clustering**

```python
clusters = {}
visited = set()

for i, m in enumerate(minhashes):
    if i in visited:
        continue
    cluster_keys = lsh.query(m) #### (4) Group items that hash into the same bucket; then query to find candidate neighbors
    cluster_indices = [int(k.split('_')[1]) for k in cluster_keys]
    for idx in cluster_indices:
        visited.add(idx)
    clusters[i] = cluster_indices

print(f"Total clusters formed: {len(clusters)}")
```

**Loading the embedding model**


```python
model = SentenceTransformer('all-MiniLM-L6-v2')
log_embeddings = model.encode(logs, show_progress_bar=True)
print("Embeddings computed.")
```

**Select Representative Events and Store them**

```python
representatives = []

for cluster_indices in clusters.values():
    cluster_vectors = log_embeddings[cluster_indices]
    similarities = cosine_similarity(cluster_vectors)
    central_idx = cluster_indices[np.argmax(similarities.sum(axis=1))]
    representatives.append({
        'RepresentativeLog': df['Content'].iloc[central_idx],
        'EventTemplate': df['EventTemplate'].iloc[central_idx]
    })

rep_df = pd.DataFrame(representatives)

rep_df.to_csv("Zookeeper_representative_logs_MaxSample.csv", index=False)
print(f"Selected {len(representatives)} representative logs. Saved to 'representative_logs.csv'.")

```
